## Importing Libs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Dummy Data Set

```python
sells = np.random.randint(100, 300, size = (12))
price = np.random.randint(25, 30, size = (12))
date = pd.date_range(start = '1/1/2021', periods = 12, freq = 'M')
```

```python
data = pd.DataFrame({
    'Date': date,
    'Price': price,
    'Sells': sells
})
```

```python
data.to_csv('stock-analysis.csv', index = False)
```

## Setting Up Data

```python
main = pd.read_csv('stock-analysis.csv')
main
```

```python
main['Earnings'] = main['Price'] * main['Sells']
main['Profits'] = main['Earnings'] * 0.12
```

```python
main
```

## Visualising the Data

### Setting Up data for ploting

```python
Date = main['Date']
Price = main['Price']
Sells = main['Sells']
Earnings = main['Earnings']
Profits = main['Profits']
plt.style.use('seaborn-whitegrid')
```

## Stock Prices Plot

```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, 
                       ncols = 2,
                       figsize = (30, 6))

price = ax0.bar(Date, Price)

fig.suptitle('Stock Prices of ABC inc.', fontweight = 'bold', fontsize = '20')

ax0.set(
    title = 'Stock Price',
    xlabel = 'Year',
    ylabel = 'Price',
    ylim = (20, 30)
);

price = ax1.plot(Date, Price)

ax1.set(
    title = 'Stock Price',
    xlabel = 'Year',
    ylabel = 'Price',
    ylim = (20, 30)
);

fig.savefig('reports/stock-prices.png')
```

## Stock Sells Plot

```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, 
                       ncols = 2,
                       figsize = (30, 6))

sells = ax0.bar(Date, Sells)

fig.suptitle('Stock Sells of ABC inc.', fontweight = 'bold', fontsize = '20')

ax0.set(
    title = 'Stock Sells',
    xlabel = 'Year',
    ylabel = 'Sells',
    ylim = (100, 300)
);

price = ax1.plot(Date, Sells)

ax1.set(
    title = 'Stock Sells',
    xlabel = 'Year',
    ylabel = 'Sells',
    ylim = (100, 300)
);

fig.savefig('reports/stock-sells.png')
```

## Earnings Plot

```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, 
                       ncols = 2,
                       figsize = (30, 6))

earnings = ax0.bar(Date, Earnings)

fig.suptitle('Earnings of ABC inc.', fontweight = 'bold', fontsize = '20')

ax0.set(
    title = 'Earnings',
    xlabel = 'Year',
    ylabel = 'Earnings',
    ylim = (2000, 9000)
);

price = ax1.plot(Date, Earnings)

ax1.set(
    title = 'Earnings',
    xlabel = 'Year',
    ylabel = 'Earnings',
    ylim = (2000, 9000)
);

fig.savefig('reports/earnings.png')
```

## Profits Plot

```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, 
                       ncols = 2,
                       figsize = (30, 6))

profits = ax0.bar(Date, Profits)

fig.suptitle('Profits of ABC inc.', fontweight = 'bold', fontsize = '20')

ax0.set(
    title = 'Profits',
    xlabel = 'Year',
    ylabel = 'Profits',
    ylim = (200, 1000)
);

price = ax1.plot(Date, Profits)

ax1.set(
    title = 'Profits',
    xlabel = 'Year',
    ylabel = 'Profits',
    ylim = (200, 1000)
);

fig.savefig('reports/profits.png')
```

## Complete Analysis Plot (BAR)

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows = 2,
                                         ncols = 2,
                                         figsize = (35, 15))

price = ax0.bar(Date, Price)
sells = ax1.bar(Date, Sells)
earnings = ax2.bar(Date, Earnings)
profits = ax3.bar(Date, Profits)

fig.suptitle('Stock Analysis of ABC inc.', fontweight = 'bold', fontsize = '20')

ax0.set(
    title = 'Stock Price',
    xlabel = 'Year',
    ylabel = 'Price',
    ylim = (20, 30)
);

ax1.set(
    title = 'Stock Sells',
    xlabel = 'Year',
    ylabel = 'Sells',
    ylim = (100, 300)
);

ax2.set(
    title = 'Earnings',
    xlabel = 'Year',
    ylabel = 'Earnings',
    ylim = (2000, 9000)
);

ax3.set(
    title = 'Profits',
    xlabel = 'Year',
    ylabel = 'Profits',
    ylim = (200, 1000)
);

fig.savefig('reports/complete-analysis-bar1.png')
```
